```python
import pandas as pd
from sklearn.metrics import mean_absolute_error

data = pd.read_csv("cvx.csv")

# Fill NaN values with 0
data.fillna(0, inplace=True)

# Convert the columns to the appropriate data type (float)
data["Actual Price"] = pd.to_numeric(data["Actual Price"], errors="coerce")
data["LSTM Model"] = pd.to_numeric(data["LSTM Model"], errors="coerce")
data["CNN Model"] = pd.to_numeric(data["CNN Model"], errors="coerce")
data["RNN Model"] = pd.to_numeric(data["RNN Model"], errors="coerce")
data["SVM Model"] = pd.to_numeric(data["SVM Model"], errors="coerce")
data["RFR Model"] = pd.to_numeric(data["RFR Model"], errors="coerce")


# Calculate the Mean Absolute Error (MAE) for each model
mae_lstm = mean_absolute_error(data["Actual Price"], data["LSTM Model"])
mae_cnn = mean_absolute_error(data["Actual Price"], data["CNN Model"])
mae_rnn = mean_absolute_error(data["Actual Price"], data["RNN Model"])
mae_svm = mean_absolute_error(data["Actual Price"], data["SVM Model"])
mae_rfr = mean_absolute_error(data["Actual Price"], data["RFR Model"])

print("CVX - Mean Absolute Error for LSTM Model: ", mae_lstm)
print("CVX - Mean Absolute Error for CNN Model: ", mae_cnn)
print("CVX - Mean Absolute Error for RNN Model: ", mae_rnn)
print("CVX - Mean Absolute Error for SVM Model: ", mae_svm)
print("CVX - Mean Absolute Error for RFR Model: ", mae_rfr)
```

```python
import matplotlib.pyplot as plt

# Define the model names and their corresponding mean absolute errors
models = ['LSTM Model', 'CNN Model', 'RNN Model', 'SVM Model', 'RFR Model']
mae = [1.7497142857142842, 2.157857142857142, 1.8482857142857134, 2.414571428571428, 2.188142857142857]

# Create a bar plot
plt.bar(models, mae)

# Set the chart title and axes labels
plt.title('Chevron - Model Mean Absolute Error')
plt.xlabel('Model')
plt.ylabel('Mean Absolute Error')

# Display the plot
plt.show()
```
