```python
import pandas as pd
from sklearn.metrics import mean_squared_error

data = pd.read_csv("mrk.csv")

# Fill NaN values with 0
data.fillna(0, inplace=True)

# Convert the columns to the appropriate data type (float)
data["Actual Price"] = pd.to_numeric(data["Actual Price"], errors="coerce")
data["LSTM Model"] = pd.to_numeric(data["LSTM Model"], errors="coerce")
data["CNN Model"] = pd.to_numeric(data["CNN Model"], errors="coerce")
data["RNN Model"] = pd.to_numeric(data["RNN Model"], errors="coerce")
data["SVM Model"] = pd.to_numeric(data["SVM Model"], errors="coerce")
data["RFR Model"] = pd.to_numeric(data["RFR Model"], errors="coerce")


# Calculate the Mean Squared Error (MSE) for each model
mse_lstm = mean_squared_error(data["Actual Price"], data["LSTM Model"])
mse_cnn = mean_squared_error(data["Actual Price"], data["CNN Model"])
mse_rnn = mean_squared_error(data["Actual Price"], data["RNN Model"])
mse_svm = mean_squared_error(data["Actual Price"], data["SVM Model"])
mse_rfr = mean_squared_error(data["Actual Price"], data["RFR Model"])

print("MRK - Mean Squared Error for LSTM Model: ", mse_lstm)
print("MRK - Mean Squared Error for CNN Model: ", mse_cnn)
print("MRK - Mean Squared Error for RNN Model: ", mse_rnn)
print("MRK - Mean Squared Error for SVM Model: ", mse_svm)
print("MRK - Mean Squared Error for RFR Model: ", mse_rfr)
```

```python
import pandas as pd
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

data = pd.read_csv("mrk.csv")

# Fill NaN values with 0
data.fillna(0, inplace=True)

# Convert the columns to the appropriate data type (float)
data["Actual Price"] = pd.to_numeric(data["Actual Price"], errors="coerce")
data["LSTM Model"] = pd.to_numeric(data["LSTM Model"], errors="coerce")
data["CNN Model"] = pd.to_numeric(data["CNN Model"], errors="coerce")
data["RNN Model"] = pd.to_numeric(data["RNN Model"], errors="coerce")
data["SVM Model"] = pd.to_numeric(data["SVM Model"], errors="coerce")
data["RFR Model"] = pd.to_numeric(data["RFR Model"], errors="coerce")

# Calculate the Mean Squared Error (MSE) for each model
mse_lstm = mean_squared_error(data["Actual Price"], data["LSTM Model"])
mse_cnn = mean_squared_error(data["Actual Price"], data["CNN Model"])
mse_rnn = mean_squared_error(data["Actual Price"], data["RNN Model"])
mse_svm = mean_squared_error(data["Actual Price"], data["SVM Model"])
mse_rfr = mean_squared_error(data["Actual Price"], data["RFR Model"])

# Define the model names and their corresponding mean squared errors
models = ['LSTM Model', 'CNN Model', 'RNN Model', 'SVM Model', 'RFR Model']
mse = [mse_lstm, mse_cnn, mse_rnn, mse_svm, mse_rfr]

# Create a bar plot
plt.bar(models, mse)

# Set the chart title and axes labels
plt.title('MRK - Model Mean Squared Error')
plt.xlabel('Model')
plt.ylabel('Mean Squared Error')

# Display the plot
plt.show()
```
