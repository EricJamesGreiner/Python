```python
import pandas as pd
from sklearn.metrics import r2_score

data = pd.read_csv("cvx.csv")

# Fill NaN values with 0
data.fillna(0, inplace=True)

# Convert the columns to the appropriate data type (float)
data["Actual Price"] = pd.to_numeric(data["Actual Price"], errors="coerce")
data["LSTM Model"] = pd.to_numeric(data["LSTM Model"], errors="coerce")
data["CNN Model"] = pd.to_numeric(data["CNN Model"], errors="coerce")
data["RNN Model"] = pd.to_numeric(data["RNN Model"], errors="coerce")
data["SVM Model"] = pd.to_numeric(data["SVM Model"], errors="coerce")
data["RFR Model"] = pd.to_numeric(data["RFR Model"], errors="coerce")


# Calculate the coefficient of determination (R-squared) for each model
r2_lstm = r2_score(data["Actual Price"], data["LSTM Model"])
r2_cnn = r2_score(data["Actual Price"], data["CNN Model"])
r2_rnn = r2_score(data["Actual Price"], data["RNN Model"])
r2_svm = r2_score(data["Actual Price"], data["SVM Model"])
r2_rfr = r2_score(data["Actual Price"], data["RFR Model"])

print("CVX - Coefficient of Determination (R-squared) for LSTM Model: ", r2_lstm)
print("CVX - Coefficient of Determination (R-squared) for CNN Model: ", r2_cnn)
print("CVX - Coefficient of Determination (R-squared) for RNN Model: ", r2_rnn)
print("CVX - Coefficient of Determination (R-squared) for SVM Model: ", r2_svm)
print("CVX - Coefficient of Determination (R-squared) for RFR Model: ", r2_rfr)
```

```python
import matplotlib.pyplot as plt

# Define the model names and their corresponding R-squared values
models = ['LSTM Model', 'CNN Model', 'RNN Model', 'SVM Model', 'RFR Model']
r_squared = [0.953, 0.941, 0.948, 0.917, 0.927]

# Create a bar plot
plt.bar(models, r_squared)

# Set the chart title and axes labels
plt.title('PG - Model R-Squared')
plt.xlabel('Model')
plt.ylabel('R-Squared')

# Display the plot
plt.show()
```
